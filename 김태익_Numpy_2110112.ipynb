```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot( h,w, maxit=20 ):
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z= c
    divtime = maxit + np.zeros(z.shape, dtype=int)
    
    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2            # who is diverging 누가 발산하나?
        div_now = diverge & (divtime==maxit)  # who is diverging now 누가 지금 발산하나?
        divtime[div_now] = i                  # note when 언제?
        z[diverge] = 2                        # avoid diverging too much 발산 너무 하는 것 피하기
    return divtime
plt.imshow(mandelbrot(400,400))
```

```python
def ufunc_reduce(ufct, *vectors):
    vs = np.ix_(*vectors)
    r = ufct.identity
    for v in vs:
        r = ufct(r,v)
    return r
```
