### 유방암 진단 분류 모델 만들기

```python
url ="https://han.gl/zNMZu"
breast_cancer = pd.read_csv(url, header=None)
breast_cancer.head()
```

```python
breast_cancer.info()
```

```python
breast_cancer.isnull().sum()
```

```python
breast_cancer.isnull().values.sum()
```

```python
breast_cancer.columns = ["id_number", "clump_thickness","unif_cell_size","unif_cell_shape","merg_adhesion",
                        "single_epith_cell_size","bare_nuclei","bland_chromatin","normal_nucleoli","mitoses","class"]
```

```python
breast_cancer
```

```python
breast_cancer['bare_nuclei'] = breast_cancer['bare_nuclei'].replace('?', np.NAN)
breast_cancer['bare_nuclei'] = \
breast_cancer['bare_nuclei'].fillna(breast_cancer['bare_nuclei'].value_counts().index[0])
```

```python
breast_cancer['cancer_ind'] = 0
breast_cancer.loc[breast_cancer['class']==4, 'cancer_ind'] = 1
```

```python
X = breast_cancer.drop(["id_number", "class", "cancer_ind"], axis=1)
y = breast_cancer.cancer_ind
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#####  KNeighborsClassifier를 이용한 학습

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_scaled, y_train)
```

```python
knn.get_params()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
y_pred = knn.predict(X_test_scaled)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(knn,{"n_neighbors":[1,2,3,4,5]}, \
                          n_jobs= -1, cv=7, scoring ="roc_auc")
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
knn_best = grid_search.best_estimator_
```

```python
y_pred = knn_best.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
roc_auc_score(y_test, y_pred)
```
